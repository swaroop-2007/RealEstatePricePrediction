```python
df1.total_sqft.unique()
```

```python
def isFloat(x):
    try:
        float(x)
    except:
        return False 
    
    return True
```

```python
df1[~df1.total_sqft.apply(isFloat)].head(15)
```

```python
def sqftAvg(x):
    tokens = x.split('-')     # Doing this for the range of sqfts
    if len(tokens) == 2:
        return (float(tokens[0]) + float(tokens[1]) ) / 2
    
    try:
        return float(x)
    except:
        return None
    
```

```python
df2 = df1.copy()

df2['total_sqft'] = df2['total_sqft'].apply(sqftAvg)
df2.head(15)
```
